```python
import numpy as np
```

```python
# create a 
df_new['Credit above limit / Credit under limit'] = np.where(df_new['AMT_CREDIT'] < df_new['AMT_APPLICATION'], 'Credit under limit', 'Credit above limit')
```

```python
df_new.head()
```

```python
# creating a new column for credit limit breached
df_new['Limit breached/Under the limit'] = df_new['AMT_CREDIT'].apply(lambda x: 'Limit breached' if x >= 350000.0 else 'Under the limit')
```

```python
df_new
```

```python
df_new.info()
```

```python
limit_breach = df_new.groupby('Limit breached/Under the limit')['NAME_CONTRACT_TYPE'].count()
limit_breach
```

```python
# Get the percentage of limit breach
Perc_limit = df_new['Limit breached/Under the limit'].value_counts()
Percentage_limit = 100 * Perc_limit / len(df_new)
print(Percentage_limit)
```

83.9% is credit under the limit and 16% above the limit, which is the firm's risk appetite.
